```python
for index, item in enumerate(img_list):
    if item:
        try:
            html = requests.get(item, verify = False)
            img_name = folder_path + height_lists[index] + "-" + name_lists[index] + '.png'
            with open(img_name, 'wb') as file:
                file.write(html.content)
                file.flush()
            file.close()
            time.sleep(1)
        except:
            pass
```
